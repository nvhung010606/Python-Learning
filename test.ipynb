```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
```

```python
tips
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=tips, x="sex", y="tip")
plt.title("Relationships between Tips and Sex")
plt.xlabel("Sex")
plt.ylabel("Tips")
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.legend(title="Day")
plt.title("Total Bill vs. Tip, Colored by Day of the Week")
plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.show()

tips['tip_percentage'] = tips['tip']/tips['total_bill']
sns.barplot(data=tips, x="day", y="tip_percentage", hue="smoker")
plt.title("Average Tip Percentage by Day and Smoker Status")
plt.show()
```
